```python
import pandas as pd

# Load the Excel file
df = pd.read_excel('../data/BondRatings.xls')

year_cols = [col for col in df.columns if col.startswith('20')]

df = df.melt(
    id_vars=['DOR CODE', 'Name', 'Rating Agency'],
    value_vars=year_cols,
    var_name='year',
    value_name='rating'
).sort_values(['Name', 'year']).reset_index(drop=True)

df['year'] = df['year'].astype(int)
display(df)
```

```python
# Create table in PostgreSQL
from sqlalchemy import create_engine, text
import os

# Build a SQLAlchemy engine for PostgreSQL
engine = create_engine(f"postgresql+psycopg2://postgres:{os.getenv('POSTGRES_PASSWORD')}@localhost:5432/postgres")

with engine.begin() as sql_conn:
	sql_conn.execute(text("CREATE SCHEMA IF NOT EXISTS bonds"))
	df.to_sql('bond_ratings', sql_conn, schema='bonds', if_exists='replace', index=False)

print(f"Loaded {len(df)} rows into bonds.bond_ratings table")
```
